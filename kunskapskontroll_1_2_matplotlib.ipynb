### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

### Answer

You can use the object oriented approach, where you explicitly create figures and axes objects and call their methods, or you could use a more direct approach where you rely on pyplot to create and manage figures and axes. Let's illustrate this by an example:

```python
domain = np.linspace(-10,10)
func_values = domain ** 3
# Object oriented approach (recommended approach)
fig, ax = plt.subplots()
ax.plot(domain,func_values)
ax.set_title(r'A plot of the function $f(x)=x^3$')
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$f(x)$')
# Direct approach
plt.figure()
plt.plot(domain,func_values)
plt.title(r'A plot of the function $f(x)=x^3$')
plt.xlabel(r'$x$')
plt.ylabel(r'$f(x)$')
plt.show()
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

### Answer

#### Figure
A figure object can be seen as the canvas on which we are about to paint our plots. It keeps track of axes, titles, legends and so on.
#### Artist
An artist is anything that is visible in a figure. This includes the figure itself, axes, titles, legends and so on. Most artists, but not all, are tied to an axes
#### Axes
An Axes object is an object that is attached to a figure and contains a region for plotting data. It contains Axis objects (as many as the dimension of the thing to be plotted), a title object and label objects for the axis objects
#### Axis
Axis objects set the scale and limits of that which is to be plotted. You can think of it as a cooordinate axis. You can use this object to control ticks (marks) on the axis and also the labels of the ticks.

### When plotting in Matplotlib, what is the expected input data type?

### Answer

The expected datatype is a numpy array but it is possible to use other data types as well with the caveat that they might not work as intended. It is recommended to convert to numpy arrays.

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
domain = np.linspace(-4,4)
func_values = domain ** 2
# Object oriented approach (recommended approach)
fig, ax = plt.subplots()
ax.plot(domain,func_values)
ax.set_title(r'A plot of the function $f(x)=x^2$')
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$f(x)$')
# Direct approach
plt.figure()
plt.plot(domain,func_values)
plt.title(r'A plot of the function $f(x)=x^2$')
plt.xlabel(r'$x$')
plt.ylabel(r'$f(x)$')
plt.show()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
fig, axs = plt.subplots(1,2,figsize=(10,5))
axs[0].scatter(random_data_x,random_data_y)
axs[0].scatter(x,y,c='red')
# Converting to numpy arrays, which is not necessary but recommended
axs[1].bar(np.array(names),np.array(values))
```
